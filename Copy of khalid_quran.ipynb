```
import pandas as pd
import numpy as np
import tqdm
import pickle

!pip install --upgrade "ibm-watson>=4.4.0"

from ibm_watson import ToneAnalyzerV3
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```
combined = pd.read_csv('English102.csv', encoding= 'unicode_escape', header = None)[0]
data = pd.DataFrame(np.concatenate(combined.str.split('|').values).reshape(-1, 3), columns = ['chapter', 'lines', 'text'])
```

```
authenticator = IAMAuthenticator('ibM5gxv4EM_1QW-lpcKh1DgRmathFlHO6JWMqSeFXI19')
tone_analyzer = ToneAnalyzerV3(
    version='2016-02-11',
    authenticator=authenticator
)

tone_analyzer.set_service_url('https://api.us-south.tone-analyzer.watson.cloud.ibm.com/instances/699ecd06-34b8-4b9e-86c1-f88fc45fc26a')

```

```
def get_info(x):
  tone_analysis = tone_analyzer.tone(x, content_type='text/plain')
  return [{j['tone_name'] : j['score'] for j in i['tones']} for i in tone_analysis.result['document_tone']['tone_categories']]
```

```
quran_info102 = [get_info(line) for line in tqdm.tqdm(data.text.values)]
```

```
# Saving the information

with open('quran_info102.pickle', 'wb') as f:
  pickle.dump(quran_info102, f, pickle.HIGHEST_PROTOCOL)
```

```
# Loading the information

with open('quran_info102.pickle', 'rb') as f:
  quran_info102 = pickle.load(f)
```
